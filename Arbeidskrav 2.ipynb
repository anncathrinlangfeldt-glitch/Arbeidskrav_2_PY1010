```python
#Oppgave 5 - Funksjon for å beregne omkrets og areal

import numpy as np

def figur_areal_omkrets(a, b):
    #Arealutregninger
    areal_halvsirkel = (np.pi * (a/2)**2)/2 #Arealet av halvsirkel
    areal_trekant = (a*b)/2 #Arealet av trekant
    areal = areal_halvsirkel + areal_trekant #Samlet areal

    #Omkretsutregninger
    omkrets_halvsirkel = np.pi * (a/2) #Omkrets av halvsirkel
    omkrets_hypotenus = np.sqrt(a**2 + b**2) #omkrets av hypotenus
    omkrets = omkrets_halvsirkel + omkrets_hypotenus + b
    
    return (areal, omkrets) #definer hva funksjonen skal retunere


a = 80 #definer verdier til a linje
b = 100 #definer verdier på b linje

areal, omkrets = figur_areal_omkrets (a, b) 

print("Areal", areal)
print ("Omkrets", omkrets)
```

```python
#Oppgave 6 - plott av mattefunksjon f(x) = x² - 5

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 200) #definer x-verdier
y= x**2 - 5 #definer y/f(x)-verdier

#lage figur
plt.figure(1)
plt.plot(x, y, color="pink", linestyle="-", linewidth = 2)
plt.grid ()
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("f(x) = x² - 5")
plt.show ()
```
